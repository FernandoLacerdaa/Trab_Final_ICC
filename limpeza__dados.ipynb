## Importar bibliotecas para tratar os dados

```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
import numpy as np
```

```python
trabalho = pd.read_csv("prin.csv")
```

### Retiramos colunas irrelevantes ou com falta de informacoes para a analise

```python
trabalho.drop(columns=["school","schoolsup","paid","famsize","Pstatus","Mjob","Fjob","higher","failures","famrel","Walc", "guardian","reason","nursery"], inplace=True)
```

```python
trabalho.info()
```

### Criando nova coluna com a media dos tres exames


```python
trabalho["media_notas"] = (trabalho["G1"] + trabalho["G3"]+trabalho["G2"])/3
```

```python
trabalho["media_notas"]
```

```python
trabalho.drop(columns=["G1","G2","G3"], inplace=True)
```

### Renomeando a escolaridade dos pais para strings

```python
# Definir o mapeamento
mapeamento_Medu = {
    0: 'analfabeto',
    1: 'primario',
    2: 'fundamental',
    3: 'medio',
    4: 'superior'
}
# Aplicar o mapeamento
trabalho['Medu'] = trabalho['Medu'].map(mapeamento_Medu)

mapeamento_Fedu = {
    0: 'analfabeto',
    1: 'primario',
    2: 'fundamental',
    3: 'medio',
    4: 'superior'
}
# Aplicar o mapeamento
trabalho['Fedu'] = trabalho['Fedu'].map(mapeamento_Fedu)
```

### Removendo observaçoes de pais com escolaridade: analfabetos
### em razao de poucas observacoes que destorceriam a analise

```python
trabalho = trabalho[trabalho['Medu'] != 'analfabeto']
trabalho = trabalho[trabalho['Fedu'] != 'analfabeto']
```

### criando o dataset final

```python
final = trabalho[["sex", "studytime", "Dalc", "freetime", "internet", "Medu", "Fedu", "media_notas"]]
```

### renomeando colunas para facilitar ao usuário


```python
final["Sexo"] = final["sex"] 
final["TempoDeEstudo"] = final["studytime"] 
final["ConsumoDeAlcool"] = final["Dalc"] 
final["TempoLivre"] = final["freetime"] 
final["Internet"] = final["internet"] 
final["EscolaridadeDaMae"] = final["Medu"] 
final["EscolaridadeDoPai"] = final["Fedu"] 
final["MediaDeNotas"] = final["media_notas"] 
```

```python
final.drop(columns = ["Dalc","freetime","internet","studytime","sex","media_notas","Medu","Fedu"],inplace=True)
```

```python
final.to_csv("student_data.csv", index=False)
```
